```python
class TreeNode(object): 
    def __init__(self, val): 
        self.val = val 
        self.left = None
        self.right = None
        self.hoehe = 0
        
    def display(self):
        lines, *_ = self._display_aux()
        for line in lines:
            print(line)

    def _display_aux(self):
        """Returns list of strings, width, height, and horizontal coordinate of the root."""
        # No child.
        if self.right is None and self.left is None:
            line = '%s' % self.val
            width = len(line)
            height = 1
            middle = width // 2
            return [line], width, height, middle

        # Only left child.
        if self.right is None:
            lines, n, p, x = self.left._display_aux()
            s = '%s' % self.val
            u = len(s)
            first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
            second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
            shifted_lines = [line + u * ' ' for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2

        # Only right child.
        if self.left is None:
            lines, n, p, x = self.right._display_aux()
            s = '%s' % self.val
            u = len(s)
            first_line = s + x * '_' + (n - x) * ' '
            second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
            shifted_lines = [u * ' ' + line for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2

        # Two children.
        left, n, p, x = self.left._display_aux()
        right, m, q, y = self.right._display_aux()
        s = '%s' % self.val
        u = len(s)
        first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
        second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
        if p < q:
            left += [n * ' '] * (q - p)
        elif q < p:
            right += [m * ' '] * (p - q)
        zipped_lines = zip(left, right)
        lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
        return lines, n + m + u, max(p, q) + 2, n + u // 2
    
    def insert (self, key):
        # binary search to find the position (recursive)
        if self.val==key:
            print("Wert schon steht")
        else:
            if self.val<key:
                if self.right==None:
                    self.right=TreeNode(key)
                else:
                    self.right.insert(key)
            else: 
                if self.left==None:
                    self.left=TreeNode(key)
                else:
                    self.left.insert(key)
        # Recalculate height
        if self.right==None:
            if self.left!=None:
                self.hoehe=self.left.hoehe+1
        else:
            if self.left!=None:
                self.hoehe=max(self.right.hoehe,self.left.hoehe)+1
            else:
                self.hoehe=self.right.hoehe+1
        #Calculate balance factor and rotation
        if self.getBalance(self.val)<-1:
            if self.getBalance(self.left.val)!=1:          #1.Fall  R
                self.rightRotate(self.val)
            else:
                self.leftRotate(self.left.val)           #3.Fall  LR 
                self.rightRotate(self.val)
        if self.getBalance(self.val)>1:
            if self.getBalance(self.right.val)!=-1:         #2.Fall L
                self.leftRotate(self.val)
            else:
                self.rightRotate(self.right.val)          #4.Fall RL
                self.leftRotate(self.val)

    def leftRotate(self, z): 
  
       #Pointer vertauschen
        if self.val==z:
            supTree=self.left
            self.left=TreeNode(self.val)
            self.left.left=supTree
            self.left.right=self.right.left
            self.val=self.right.val
            self.right=self.right.right
        else:
            if self.val<z:
                self.right.leftRotate(z)
            else:
                self.left.leftRotate(z)
    
       #Recalculate height
        if self.left.right==None:
            if self.left.left!=None:
                self.left.hoehe=self.left.left.hoehe+1
        else:
            if self.left.left!=None:
                self.left.hoehe=max(self.left.right.hoehe,self.left.left.hoehe)+1
            else:
                self.left.hoehe=self.left.right.hoehe+1
        if self.right==None:
            if self.left!=None:
                self.hoehe=self.left.hoehe+1
        else:
            if self.left!=None:
                self.hoehe=max(self.right.hoehe,self.left.hoehe)+1
            else:
                self.hoehe=self.right.hoehe+1

    def rightRotate(self, z): 
      
       #Pointer vertauschen               
        if self.val==z:
            supTree=self.right
            self.right=TreeNode(self.val)
            self.right.right=supTree
            self.right.left=self.left.right
            self.val=self.left.val
            self.left=self.left.left
            del supTree
        else:
            if self.val<z:
                return self.right.rightRotate(z)
            else:
                return self.left.rightRotate(z)
    
       #Recalculate height
        if self.right.right==None:
            if self.right.left!=None:
                self.right.hoehe=self.right.left.hoehe+1
        else:
            if self.right.left!=None:
                self.right.hoehe=max(self.right.right.hoehe,self.right.left.hoehe)+1
            else:
                self.right.hoehe=self.right.right.hoehe+1
        if self.right==None:
            if self.left!=None:
                self.hoehe=self.left.hoehe+1
        else:
            if self.left!=None:
                self.hoehe=max(self.right.hoehe,self.left.hoehe)+1
            else:
                self.hoehe=self.right.hoehe+1


    def getHeight(self, root): 
        # zurückgeben der Variable Höhe für root.
        # root finden:
        if self.val==root:
            return self.hoehe
        else:
            if self.val < root:
                return self.right.getHeight(root)
            else:
                return self.left.getHeight(root)
            
    def getBalance(self, root): 
        if self.val==root:
            if self.right==None:
                if self.left==None:
                    return 0
                else:
                    return 0-self.getHeight(self.left.val)-1
            else:
                if self.left==None:
                    return self.getHeight(self.right.val)+1
                else:
                    return self.getHeight(self.right.val)-self.getHeight(self.left.val)
        else:
            if self.val<root:
                return self.right.getBalance(root)
            else:
                return self.left.getBalance(root)
            
    def remove(self,root):
        if self.val==root:
            if self.left!=None:
                self.val=self.left.getMax()
                if self.left.val==self.val:
                    if self.left.left==None:
                        del self.left
                        self.left=None
                    else:
                        self.left.val=self.left.left.val
                        del self.left.left
                        self.left.left=None
                else:
                    self.left.delTreeNode(self.val)
            else:
                self.val=self.right.val
                del self.right
                self.right=None
        else:
            if self.right!=None:
                if self.right.val==root:
                    if self.right.right==None:
                        if self.right.left==None:
                            del self.right
                            self.right=None
                        else:
                            self.right.val=self.right.left.val
                            del self.right.left
                            self.right.left=None
                    else:
                        if self.right.left==None:
                            self.right.val=self.right.right.val
                            del self.right.right
                            self.right.right=None
                        else:
                            self.right.val=self.right.left.getMax()
                            if self.right.left.val==self.right.val:
                                if self.right.left.left==None:
                                    del self.right.left
                                    self.right.left=None
                                else:
                                    self.right.left.val=self.right.left.left.val
                                    del self.right.left.left
                                    self.right.left.left=None
                            else:
  
                                self.right.left.delTreeNode(self.right.val)

            if self.left!=None:
                if self.left.val==root:
                    if self.left.right==None:
                        if self.left.left==None:
                            del self.left
                            self.left=None
                        else:
                            self.left.val=self.left.left.val
                            del self.left.left
                            self.left.left=None
                    else:
                        if self.left.left==None:
                            self.left.val=self.left.right.val
                            del self.left.right
                            self.left.right=None
                        else:
                            self.left.val=self.left.left.getMax()
                            if self.left.left.val==self.left.val:
                                if self.left.left.left==None:
                                    del self.left.left
                                    self.left.left=None
                                else:
                                    self.left.left.val=self.left.left.left.val
                                    del self.left.left.left
                                    self.left.left.left=None
                            else:
                                self.left.left.delTreeNode(self.left.val)
            if self.val<root:
                if self.right!=None:
                    if self.right.val<root:
                        if self.right.right!=None:
                             self.right.right.remove(root)
                    else:
                        if self.right.left!=None:
                             self.right.left.remove(root)
            else:
                if self.left!=None:
                    if self.left.val<root:
                        if self.left.right!=None:
                             self.left.right.remove(root)
                    else:
                        if self.left.left!=None:
                             self.left.left.remove(root)
        self.getNewHeightandNewB()
            
    def getMax(self):
        if self.right!=None:
            return self.right.getMax()
        else:
            return self.val
        
    def delTreeNode(self,root):
        if self.right.val==root:
            del self.right
            self.right=None
        else:
            if self.val<root:
                self.right.delTreeNode(root)
            else:
                self.left.delTreeNode(root)
        self.getNewHeightandNewB()
            
    def getNewHeightandNewB(self):
        if self.right==None:
            if self.left!=None:
                self.hoehe=self.left.hoehe+1
        else:
            if self.left!=None:
                self.hoehe=max(self.right.hoehe,self.left.hoehe)+1
            else:
                self.hoehe=self.right.hoehe+1
        #Calculate balance factor and rotation
        if self.getBalance(self.val)<-1:
            if self.getBalance(self.left.val)!=1:          #1.Fall  R
                self.rightRotate(self.val)
            else:
                self.leftRotate(self.left.val)           #3.Fall  LR 
                self.rightRotate(self.val)
        if self.getBalance(self.val)>1:
            if self.getBalance(self.right.val)!=-1:         #2.Fall L
                self.leftRotate(self.val)
            else:
                self.rightRotate(self.right.val)          #4.Fall RL
                self.leftRotate(self.val)
```
